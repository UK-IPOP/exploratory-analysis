# Synthetic Data Creation Practice

```
# import from standard library
import random
from collections import Counter
```

## 1. Simple 50/50 probability

```
# randomly choose heads or tails 1000 times and display distribution counts
data: list = []
i: int = 0
while i < 1_000:
    choice: str = random.choice(('heads', 'tails'))
    data.append(choice)
    i += 1

# use most_common for sorted results
Counter(data).most_common()
```

## 2. Three-way distribution of marbles

### 2.1 Even Distribution

```
# this set has even distribution
data: list = []
i: int = 0
while i < 1_000:
    choice: str = random.choice(("red", "green", "blue"))
    data.append(choice)
    i += 1

Counter(data).most_common()
```

```
# set up an unevent distribution
marble_distribution = {"blue": 50, "green": 30, "red": 20}

# this set is weighted
Counter(random.choices(('red', 'green', 'blue'), weights=(.20, .30, .50), k=1000)).most_common()
```
